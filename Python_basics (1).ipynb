2. **What is an interpreter in Python?**
    - An interpreter in Python is a program that reads and executes Python code line by line. Unlike compilers, which translate the entire source code into machine code at once, an interpreter processes and runs the code sequentially, one instruction at a time.

      ***some key points*** :
        -  Line-by-line execution: Python code is read, parsed, and executed line by line.
        - No separate compilation step: Python does not require compiling before running.
        - Easier debugging: Errors are shown as they occur, making it easier to identify and fix issues.
        - Interactive mode: Python’s interpreter allows running code directly in a shell or terminal.
    
     **Popular Python Interpreters:**
        - CPython: The default and most widely used implementation, written in C.
        - PyPy: A faster alternative that uses Just-In-Time (JIT) compilation.
        - Jython: Runs on the Java platform.
        - IronPython: Integrates with the .NET framework.

6. **Why are lists mutable, but tuples are immutable?**
    - lists are mutable because :
        - **Design Choice**: Lists are designed for flexibility. They're often used when you need to add, remove, or change elements.
        - **Implementation**: Internally, lists are stored in a way that allows direct access and modification of elements.
        - **Use Case**: Useful in scenarios like dynamic data handling, where the size or content of the collection may change during execution.


````python
#souce code for list

my_list = [1, 2, 3]
my_list[0] = 100       # List allows modification
my_list.append(4)      # List allows adding elements
print(my_list)         # Output: [100, 2, 3, 4]
````

  - tuples are immutable because :
      - Design Choice: Tuples are made for fixed data. They're ideal when you want to ensure data stays constant.

      - Performance: Because they can’t change, tuples can be optimized for performance and memory usage.

      - Safety: Tuples are safe to use as dictionary keys or elements in sets because their content won’t change (only immutable objects can be hashed reliably).

````python
# source code for tuples

my_tuple = (1, 2, 3)
my_tuple[0] = 100      # Error: Tuples cannot be changed
````

7. **What is the difference between “==” and “is” operators in Python?**
    -  **== (Equality Operator)**
      - Checks if two values are equal in terms of content.
      - It compares the data stored in the objects.
      - example :
      ```
      #source code
      a = [1, 2, 3]
      
      b = [1, 2, 3]
      
      print(a == b)   # True → because both lists have the same content
      ```
    -  **is (Identity Operator)**
      - Checks if two variables point to the same object in memory.
      - It compares the memory location (object identity).
      
      ```
        #source code
        a = [1, 2, 3]
        
        b = [1, 2, 3]
        
        c = a
        print(a is b)  # False → different objects with same content
        print(a is c)  # True → both point to the same object
      ```





8. **What are logical operators in Python?**
    - Logical operators are used to combine conditional statements (Boolean expressions). They return True or False based on the logic.
    - Python's 3 Logical Operators:
       - and : Returns True if both conditions are true	x > 5 and x < 10
       - or :  Returns True if at least one condition is true	x > 5 or x < 3
       - not :	Reverses the result (True ⇄ False)	not (x > 5)
          
          ```
          # Example code
          x = 7

          # AND
          print(x > 5 and x < 10)   # True (both conditions are true)

          # OR
          print(x > 10 or x == 7)   # True (second condition is true)

          # NOT
          print(not(x > 5))         # False (x > 5 is True, so not True → False)

          ```




9. **What is type casting in Python?**
    - Type casting (or type conversion) in Python is the process of converting the value of one data type into another.
    - Python supports
        - Explicit Type Casting (Manual)
        - Implicit Type Casting (Automatic)
    1. **Explicit Type Casting (Manual)**
      ```
      # This is formatted as code
      # String to int
      a = int("10")      # 10 (integer)

      # Float to int
      b = int(5.7)       # 5

      # Int to float
      c = float(3)       # 3.0

      # Number to string
      d = str(100)       # "100"

      # List from string
      e = list("abc")    # ['a', 'b', 'c']
      ```
    2. **Implicit Type Casting (Automatic)**
        ```
        # This is formatted as code
        x = 10      # int
        y = 3.5     # float
        z = x + y   # z becomes float automatically
        print(type(z))     # <class 'float'>
        ```

13. **What is the difference between for and while loops ?**
    - Both for and while loops are used for repeating a block of code, but they work differently and are used in different scenarios.
    - for Loop:
        - Used when you know in advance how many times you want to iterate.
        - Often used to loop through a sequence (like a list, string, or range).
    - while Loop:
        - Used when you don’t know how many times to loop — you loop as long as a condition is true.
        - Good for repeating until a condition changes.


````python
# example code of for loop
for i in range(5):
    print(i)
````

````python
# example code of while loop
i = 0
while i < 5:
    print(i)
    i += 1
````

6.  **Write a function to demonstrate the behavior of mutable and immutable arguments.**

````python
def modify_arguments(num, num_list):
    num += 10  # Attempt to modify the integer (immutable)
    print("Inside function - num:", num)

    num_list.append(100) # Modify the list (mutable)
    print("Inside function - num_list:", num_list)

# Original variables
x = 5
y = [1, 2, 3]

# Call the function
modify_arguments(x, y)

# Check after function call
print("Outside function - x:", x)          # Immutable → remains unchanged
print("Outside function - y:", y)          # Mutable → changed
````

8. **Write a program to demonstrate the use of logical operators.**

````python
# Take three numbers as input
a = int(input("Enter first number (a): "))
b = int(input("Enter second number (b): "))
c = int(input("Enter third number (c): "))

# Logical AND: Check if all numbers are positive
if a > 0 and b > 0 and c > 0:
    print(" All numbers are positive.")
else:
    print(" Not all numbers are positive.")

# Logical OR: Check if at least one number is zero
if a == 0 or b == 0 or c == 0:
    print(" At least one number is zero.")
else:
    print(" No number is zero.")

# Logical NOT: Check if 'a' is NOT greater than 'b'
if not a > b:
    print(" 'a' is NOT greater than 'b'")
else:
    print(" 'a' is greater than 'b'")
````

 10. **Write code to demonstrate type casting with list elements.**

````python
# List of strings representing numbers
str_list = ["10", "20", "30", "40"]

# Convert to integers using list comprehension
int_list = [int(x) for x in str_list]
print("String to Integer:", int_list)

# Convert integers to floats
float_list = [float(x) for x in int_list]
print("Integer to Float:", float_list)

# Convert floats to strings
str_again_list = [str(x) for x in float_list]
print("Float to String:", str_again_list)
````
